## Logistic Regression

### Hypothesis

$$
  P (survive | age, sex) = \sigma(ax_i + by_i +c)
$$

#### Loss Function

$L(a,b,c) = \sum_{i = 1}^n (-t_i.ln(\sigma(ax_i + by_i + cz_i + d)) - (1 - t_i).ln(1 - \sigma(ax_i + by_i +cz_i + d)))$

#### Derivatives

$\frac{\partial L}{\partial a} = \sum_{i = 1}^n -x_i(t_i - \sigma(ax_i + by_i + cz_i + d))$

$\frac{\partial L}{\partial b} = \sum_{i = 1}^n -y_i(t_i - \sigma(ax_i + by_i + cz_i + d))$

$\frac{\partial L}{\partial c} = \sum_{i = 1}^n -z_i(t_i - \sigma(ax_i + by_i + cz_i + d))$

$\frac{\partial L}{\partial d} = \sum_{i = 1}^n -(t_i - \sigma(ax_i + by_i + cz_i + d))$

### Implementation

```python
class LossFunction:
  def __init__(self, x, y, z, t):
    self.x = x
    self.y = y
    self.z = z
    self.t = t

  def sigmoid(self, x):
    return 1 / (1 + np.exp(-x))

  def value(self, a, b, c, d):
    value_array = self.sigmoid(a * self.x + b * self.y + c * self.z + d)
    loss_values = -self.t * np.log(value_array) - (1 - self.t) * np.log(1 - value_array)
    return np.nansum(loss_values)

  def derivative(self, a, b, c, d):
    prediction = self.sigmoid(a * self.x + b * self.y + c * self.z + d)
    common_part = - (self.t - prediction)
    return np.array([np.nansum(common_part * self.x), np.nansum(common_part * self.y), np.nansum(common_part * self.z), np.nansum(common_part)])

  def accuracy(self, a, b, c, d):
        predictions = self.sigmoid(a * self.x + b * self.y + c * self.z + d)
        predictions = np.array([1 if x > 0.5 else 0 for x in predictions])
        return np.sum(predictions == self.t) / len(self.t)
```

```python
func = LossFunction(sex, age, pclass, survided)
func.derivative(1, 1, 1, 1)
```

```python
def gradient_descent_process(starting_point, learning_rate, iterations):

    # clear the graphs folder
    import os
    import shutil
    shutil.rmtree('./graphs')
    os.mkdir('./graphs')

    x = np.linspace(-7, -0, 100)
    y = np.linspace(-2, 0, 100)
    X, Y = np.meshgrid(x, y)

    current_point = starting_point
    point_storage = []
    for i in range(iterations):
        plt.clf()
        Z = [[func.value(x, y, current_point[2], current_point[3]) for x in x] for y in y]
        plt.contourf(X, Y, Z, 20, cmap = 'RdGy')
        plt.title('z = ' + str(current_point[2]) + ' t = ' + str(current_point[3]))
        plt.colorbar()
        
        derivative = func.derivative(current_point[0], current_point[1], current_point[2], current_point[3])
        print("Iteration: " + str(i))
        print("Current point: " + str(current_point))
        print("derivative: " + str(derivative))
        print("Loss value: " + str(func.value(current_point[0], current_point[1], current_point[2], current_point[3])))
        print("Accuracy: " + str(func.accuracy(current_point[0], current_point[1], current_point[2], current_point[3])))
        print("====================================")
        next_point = current_point - derivative * learning_rate
        # plot the point_storage of the points as gray, the current point as red
        point_storage.append(current_point)
        for point in point_storage:
            plt.scatter(point[0], point[1], c = 'grey')
        plt.scatter(next_point[0], next_point[1], c = 'red')
        #plot the line between points in point_storage
        for i in range(len(point_storage) - 1):
            plt.plot([point_storage[i][0], point_storage[i + 1][0]], [point_storage[i][1], point_storage[i + 1][1]], c = 'yellow')
        plt.plot([current_point[0], next_point[0]], [current_point[1], next_point[1]], c = 'yellow')
        
        plt.savefig("./graphs/iteration_" + str(i) + ".png")

        current_point = next_point

        if (np.linalg.norm(derivative) < 0.00001):
            break

    return current_point

starting_point = np.array([-2.5, -0.5, -2, 2.5])
learning_rate = 0.001
iterations = 100
gradient_descent_process(starting_point, learning_rate, iterations)
```
